```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib
matplotlib.rcParams['figure.figsize'] = [12.0, 8.0]
```

```python
def plot_local_cluster(local_data_x, local_data_y, cluster_data_x=[], cluster_data_y=[]):
    plt_range_min_y = int(min(local_data_y + cluster_data_y) * 0.9)
    plt_range_max_y = int(max(local_data_y + cluster_data_y) * 1.1)
    plt_range_min_x = int(min(local_data_x + cluster_data_x) * 0.9)
    plt_range_max_x = int(max(local_data_x + cluster_data_x) * 1.1)
    plt.plot(local_data_x, local_data_y, 'b', local_data_x, local_data_y, 'bo')
    plt.plot(cluster_data_x, cluster_data_y, 'r', cluster_data_x, cluster_data_y, 'ro')
    plt.axis([plt_range_min_x, plt_range_max_x, plt_range_min_y, plt_range_max_y])
    plt.ylabel('seconds')
    plt.xlabel('MB')
    blue_patch = mpatches.Patch(color='blue', label='local')
    red_patch = mpatches.Patch(color='red', label='cluster')
    plt.legend(handles=[blue_patch, red_patch])
    plt.show()
    
def plot_local_triple_local(local_data_x, local_data_y_1, local_data_y_2, local_data_y_total, local_data_y):
    plt_range_min_y = int(min(local_data_y + local_data_y_total) * 0.9)
    plt_range_max_y = int(max(local_data_y + local_data_y_total) * 1.1)
    plt_range_min_x = int(min(local_data_x) * 0.9)
    plt_range_max_x = int(max(local_data_x) * 1.1)
    plt.plot(local_data_x, local_data_y_total, 'k', local_data_x, local_data_y_total, 'ko')
    plt.plot(local_data_x, local_data_y_1, 'y', local_data_x, local_data_y_1, 'yo')
    plt.plot(local_data_x, local_data_y_2, 'g', local_data_x, local_data_y_2, 'go')
    plt.plot(local_data_x, local_data_y, 'b', local_data_x, local_data_y, 'bo')
    plt.axis([plt_range_min_x, plt_range_max_x, plt_range_min_y, plt_range_max_y])
    plt.ylabel('seconds')
    plt.xlabel('MB')
    yellow_patch = mpatches.Patch(color='yellow', label='local - 2x MapReduce (first)')
    green_patch = mpatches.Patch(color='green', label='local - 2x MapReduce (second)')
    black_patch = mpatches.Patch(color='black', label='local - 2x MapReduce (total)')
    blue_patch = mpatches.Patch(color='blue', label='local - 1x MapReduce')
    plt.legend(handles=[yellow_patch, green_patch, black_patch, blue_patch])
    plt.show()
```

```python
print("MapReduce - Job 1")
plot_local_cluster([5, 26, 70, 184], [15, 17, 23, 31])
```

```python
print("MapReduce - Job 2")
plot_local_cluster([5, 26, 70, 184], [15, 19, 24, 42])
```

```python
print("MapReduce - Job 2, local comparison")
plot_local_triple_local([5, 26, 70, 184], [14, 15, 16, 24], [17, 20, 27, 44], [14+1+17, 15+1+20, 16+1+27, 24+1+44], [16, 17, 23, 47])
print("MapReduce - Job 2, 1x MapReduce")
plot_local_cluster([5, 26, 70, 184], [14+1+17, 15+1+20, 16+1+27, 24+1+44])
print("MapReduce - Job 2, 2x MapReduce")
plot_local_cluster([5, 26, 70, 184], [16, 17, 23, 47])
```
